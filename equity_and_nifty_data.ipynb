```python
import pandas as pd
```

```python
all_company_equity=pd.read_csv("D:\\stock_market analysis\\EQUITY_L.csv")
print(all_company_equity)
```

```python
nifty_100_companies=pd.read_csv("D:\stock_market analysis\ind_nifty100list.csv")
print(nifty_100_companies)
```

```python
detailed_nifty_100_companies=pd.merge(all_company_equity, nifty_100_companies, on=["Symbol","ISIN Code","Series"])
print(detailed_nifty_100_companies)
```

```python
columns_to_be_displayed=detailed_nifty_100_companies[['Symbol','Company Name','Series','Date of listing','ISIN Code','Industry','Face value']]
print(columns_to_be_displayed)
```

```python
sorted_data_based_on_facevalue=columns_to_be_displayed.sort_values('Face value')
print(sorted_data_based_on_facevalue)
```

```python
top_100_companies=sorted_data_based_on_facevalue.head(100)
print(top_100_companies)
```

```python
def all_data():
    print(all_company_equity)

def top_100():
    print(top_100_companies)

def search():
    symbol_name = input("Enter the symbol of the company: ")
    list_of_symbols=columns_to_be_displayed['Symbol'].tolist()
    
    if symbol_name in list_of_symbols:
        searched_data = columns_to_be_displayed[columns_to_be_displayed['Symbol'] == symbol_name]
        print(searched_data)
    else:
        print("Data not found")

flag = True
while flag == True:
    print('''
       ****** MENU *******
    Select from the following:- 
    
    1) Display all data, Press 1 
    2) Display Top 100, Press 2
    3) Search, Press 3
    4) Exit, Press 4''')

    choice = int(input("\nEnter your choice:- "))

    if choice == 1:
        all_data()

    elif choice == 2:
        top_100()

    elif choice == 3:
        search()

    elif choice == 4:
        print("Exiting, Thank you!")
        flag = False
    else:
        print("\nInvalid input")
```
