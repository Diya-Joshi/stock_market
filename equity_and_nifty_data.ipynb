```python
import pandas as pd
```

```python
all_company_equity=pd.read_csv("D:\\stock_market analysis\\EQUITY_L.csv")
print(all_company_equity)
```

```python
sorted_all_company_equity=all_company_equity.sort_values('Face value')
print(sorted_all_company_equity)
```

```python
nifty_100_companies=pd.read_csv("D:\stock_market analysis\ind_nifty100list.csv")
print(nifty_100_companies)
```

```python
detailed_nifty_100_companies=pd.merge(all_company_equity, nifty_100_companies, on=["Symbol","ISIN Code","Series"])
print(detailed_nifty_100_companies)
```

```python
top_100_companies=detailed_nifty_100_companies[['Symbol','Company Name','Series','Date of listing','ISIN Code','Industry','Face value']]
print(top_100_companies)
```

```python
sorted_top_100_companies=top_100_companies.sort_values('Face value')
print(sorted_top_100_companies)
```

```python
def all_data():
    print(all_company_equity)

def sorted_all_data():
    print(sorted_all_company_equity)

def top_100():
    print(top_100_companies)

def sorted_top_100():
    print(sorted_top_100_companies)
    
def search():
    symbol_name = input("Enter the symbol of the company: ")
    list_of_symbols=all_company_equity['Symbol'].tolist()
    
    if symbol_name in list_of_symbols:
        searched_data = all_company_equity[all_company_equity['Symbol'] == symbol_name]
        print(searched_data)
    else:
        print("Data not found")

flag = True
while flag == True:
    print('''
       ****** MENU *******
    Select from the following:- 
    
    1) Display all data, Press 1 
    2) Display Top 100, Press 2
    3) Search, Press 3
    4) Exit, Press 4''')

    choice = int(input("\nEnter your choice:- "))

    if choice == 1:
        all_data()
        sort_choice = input("\nWould you like to sort data on the basis of Face Value (y/n)?:- ")
        
        if sort_choice == 'y' or sort_choice == 'Y':
            sorted_all_data() 
            
        elif sort_choice == 'n' or sort_choice == 'N':
            pass
        
        else:
            print("Invalid input!")
            
        continue_choice=input("\nDo you want to continue (y/n)?:- ")
        
        if continue_choice == 'y' or continue_choice == 'Y':
            continue
        else:
            break

    elif choice == 2:
        top_100()
        sort_choice = input("\nWould you like to sort data on the basis of Face Value (y/n)?:- ")
        
        if sort_choice == 'y' or sort_choice == 'Y':
            sorted_top_100() 
            
        elif sort_choice == 'n' or sort_choice == 'N':
            pass
        
        else:
            print("Invalid input!")
            
        continue_choice=input("\nDo you want to continue (y/n)?:- ")
        
        if continue_choice == 'y' or continue_choice == 'Y':
            continue
        else:
            break

    elif choice == 3:
        search()
        
        continue_choice=input("\nDo you want to continue (y/n)?:- ")
        
        if continue_choice == 'y' or continue_choice == 'Y':
            continue
        else:
            break

    elif choice == 4:
        print("Exiting, Thank you!")
        flag = False
    else:
        print("\nInvalid input")
```
